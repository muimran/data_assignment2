```python
import pandas as pd
```

```python
df = pd.read_csv("dhaka_apartment_sale_2023.csv", on_bad_lines='skip')
```

```python
df
```

```python
# Convert 'size' to string, then remove commas and convert to integer
df['size'] = df['size'].astype(str).str.replace(',', '').astype(int)

# Calculate 'ppsf' and round to the nearest whole number
df['ppsf'] = (df['price'] / df['size']).round()

# Alternatively, calculate 'ppsf' and convert to integer (this truncates the decimal part)
# df['ppsf'] = (df['price'] / df['size']).astype(int)

# Display the DataFrame
df
```

```python
neighborhood = ["mohammadpur", "mirpur", "badda", "adabor", "banasree", "uttara", "bashundhara r-a", "dhanmondi", "banani", "gulshan"]

filtered_dfs = {}
for word in neighborhood:
    # Filter df where 'location' column contains the word
    filtered_dfs[word] = df[df['address'].str.contains(word, case=False, na=False)]

df
```

```python
mohammadpur = filtered_dfs['mohammadpur']
mirpur = filtered_dfs['mirpur']
badda = filtered_dfs['badda']
adabor = filtered_dfs['adabor']
banasree = filtered_dfs['banasree']
uttara = filtered_dfs['uttara']
bashundhara = filtered_dfs['bashundhara r-a']
dhanmondi = filtered_dfs['dhanmondi']
banani = filtered_dfs['banani']
gulshan = filtered_dfs['gulshan']
```
